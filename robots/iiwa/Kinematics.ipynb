```python
from sympy import *
init_session()
```

```python
q1, q2, q3, q4, q5, q6, q7 = symbols('q_1 q_2 q_3 q_4 q_5 q_6 q_7')
d_bs, d_se, d_ew, d_wf = symbols('d_bs d_se d_ew d_wf')
q = Matrix([q1, q2, q3, q4, q5, q6, q7])
dh_a = Matrix([0., 0., 0., 0., 0., 0., 0.])
dh_alpha = Matrix([-pi / 2, pi / 2, pi / 2, -pi / 2, -pi / 2, pi / 2, 0.])
dh_d = Matrix([d_bs, 0., d_se, 0., d_ew, 0., d_wf])
```

```python
def transform_i(theta_i, index):
    alpha_i = dh_alpha[index]
    a_i = dh_a[index]
    d_i = dh_d[index]
    T_i = Matrix.eye(4)
    T_i[0, 0] = cos(theta_i)
    T_i[0, 1] = -sin(theta_i) * cos(alpha_i)
    T_i[0, 2] = sin(theta_i) * sin(alpha_i)
    T_i[0, 3] = a_i * cos(theta_i)
    T_i[1, 0] = sin(theta_i)
    T_i[1, 1] = cos(theta_i) * cos(alpha_i)
    T_i[1, 2] = -cos(theta_i) * sin(alpha_i)
    T_i[1, 3] = a_i * sin(theta_i)
    T_i[2, 0] = 0.
    T_i[2, 1] = sin(alpha_i)
    T_i[2, 2] = cos(alpha_i)
    T_i[2, 3] = d_i
    return T_i
```

```python
# Forward Kinematics
T = Matrix.eye(4)
for i in range(7):
    T = T * transform_i(q[i], i)
T
```

```python
jac_1 = []
for i in range(7):
    jac_1.append(diff(T[0, 3], q[i]))
jac_1
```
